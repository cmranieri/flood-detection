```python
import pandas as pd
import numpy as np
import math
import os
import enoe_utils
import re
import shutil
```

```python
csv_path = 'flood_images_annot.csv'
df = enoe_utils.load_df(csv_path, place=None)
```

```python
df.head()
```

```python
src_dir = '/content/drive/MyDrive/enoe'
dst_dir = '/content/drive/MyDrive/enoe2'

for index, row in df.iterrows():
    path_split = os.path.split(row['path'])
    fname = path_split[1]
    mtc = re.match('(\d\d\d\d)(\d\d)(\d\d)_(\d\d)(\d\d)(\d\d)', fname)
    yyyy = mtc.groups()[0]
    mm = mtc.groups()[1]
    dd = mtc.groups()[2]
    old_path = os.path.join( src_dir, yyyy, fname )
    new_path = os.path.join( dst_dir, yyyy, mm, dd, fname )
    os.makedirs( os.path.join(dst_dir, yyyy, mm, dd) )
    shutil.copyfile(old_path, new_path)
    df.loc[ index, 'path' ] = new_path
    print( old_path, new_path )

df.to_csv(csv_path)
```
